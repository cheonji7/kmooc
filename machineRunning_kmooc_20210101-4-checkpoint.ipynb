## 2-2강 AND 게이트 뉴론 구현하기 ##

```python
## 2-2 AND 게이트 뉴론 구현하기
import numpy as np

def AND(x1, x2):
    x = np.array( [ 1, x1, x2 ] )       # input
    w = np.array( [-0.7, 0.5, 0.5])     # bia s + weight
    ## return (np.dot(w,x) > 0)         #True, False 를 리턴
    return int(np.dot(w, x) > 0)    # int  1 or  0 를 리턴


## 테스트 
print("AND(0, 0)= " + str(AND(0, 0)) )
print("AND(0, 1)= " + str(AND(0, 1)) )
print("AND(1, 0)= " + str(AND(1, 0)) )
print("AND(1, 1)= " + str(AND(1, 1)) )
```

## AND게이트 뉴론 시각화  ##
## 0.7 +0.5*x1 + 0.5*x2 = 0  ##
## x2 = -x1 + 1.4

## OR  5*x1 + 0.5*x2 = 0  ##
## x2 = -x1 + 1.4

```python
## 2-2 AND 게이트 뉴론 구현하기
import numpy as np

def OR(x1, x2):
    x = np.array( [ 1, x1, x2 ] )       # input
    w = np.array( [-0.3, 0.5, 0.5])     # bia s + weight
    ## return (np.dot(w,x) > 0)         #True, False 를 리턴
    return int(np.dot(w, x) > 0)    # int  1 or  0 를 리턴


## 테스트 
print("OR(0, 0)= " + str(OR(0, 0)) )
print("OR(0, 1)= " + str(OR(0, 1)) )
print("OR(1, 0)= " + str(OR(1, 0)) )
print("OR(1, 1)= " + str(OR(1, 1)) )
```

```python

## 체온변환
## 목적: 뉴론은 입력인자 C를 받아서 F(화씨)로 환산하여 출력
## 출력(리턴)하기 전에 뉴론은 활성화 할지 비활성화할지 activate()함수를 호출함

def activate(z):
    """retruns 32 if z < 32"""
    if z < 32:
        z = 32
    return z

def C2F(C):
    """converts celcius to Fahrenheit"""
    F= 9/5 * C + 32
    return activate(F)


test_c = [-20, -10, 0, 32, 36, 40, 50, 400]
test_f = [C2F(c) for c in test_c]
print(test_f)


import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


#Plotting the simple newron

x =np.arange(-100, 100, 0.1)
y = [C2F(i)  for i in x]

plt.figure()
plt.plot(x, y)
plt.axis([-20, 50, 0, 150])
plt.xlabel('celcius')
plt.ylabel('Fahrenheit')
plt.show()
```

## 2-4 퍼셉트론 예제 코딩 ##


```python
import numpy as np
x = np.array([1.0, 1.0], [2.0, -2.0], [-1.0,-1.5],[-2.0, -1.0], [-2.0, 1.0], [1.5, -0.5])
X = np.c_[np.ones(len(x),x)]

y=np.array([1, -1, -1, -1, 1, 1])
w=np.array([0.0, 1.0, 0.5]).reshape(X.shape[1],1)
maxlabel, minlabel = y.max(), y.min()

eta = 0.1
for xi, yi in zip(X, y):
    xi = xi.reshape(w.shape)
    z = np.dot(x.T, xi)
    yhat = np.where(z>0.0, 1, -1)
    delta = eta *(yi-yhat) *xi
    w += delta
                    
print(np.round(w, 2))
                    
                    
                    
```
